```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
```

```python
def lineChart(V,I,ID):
    '''
    V: data of the first axis
    I: data of the second axis
    ID: context of the text box
    '''
    #V= np.transpose(V)
    plt.gca().set_prop_cycle(plt.cycler('color', plt.cm.jet(np.linspace(0, 1, len(V)+len(I)))))# Plot several different functions..
    labels1 = []
    labels2 = []
    plt.title('Voltage/Current-Time', fontsize='16')  # name of chart
    plt.xlabel('Time', fontsize='12')  # name of x axis, size of word
    plt.ylabel('Voltage', fontsize='12')  # name of y axis, size of word
    plt.ylim(0, 20)
    for i in range(0, len(V) ):
        x, y = range(0, len(V[0])), V[i]  
        plt.plot(x, y)
        labels1.append(r'V:%i ' %  i)
    plt.legend(labels1, ncol=4, loc='upper right',
               bbox_to_anchor=[1, 1],
           columnspacing=1.0, labelspacing=0.2,
           handletextpad=0.2, handlelength=1.5,
           fancybox=True, shadow=True)
    
    plt.twinx()#the second axis
    for j in range(0, len(I) ):
        x, y = range(0, len(I[0])), I[j]  
        plt.plot(x, y)
        labels2.append(r'I:%i ' %  j)
    plt.ylabel('Current', fontsize=12)
    plt.ylim(0, 20)
    plt.figtext(0.3, 0.3, f"text:%s" %ID,fontsize='16')#text box
    plt.legend(labels2, ncol=4, loc='lower right',
           bbox_to_anchor=[1, 0],
           columnspacing=1.0, labelspacing=0.2,
           handletextpad=0.2, handlelength=1.5,
           fancybox=True, shadow=True)
    plt.savefig('%s.jpg' % ID)  #
    plt.show()
```

```python
V=np.random.rand(2,20)
I=np.random.randint(10,20,(2,20))
ID='test'
print(V)
print(I)
lineChart(V,I,ID)
```
